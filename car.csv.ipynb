```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
car=pd.read_csv("cars.csv")
```

```python
car.head(5)
```

```python
car.tail(5)
```

```python
car.describe(include='all')
```

```python
car.hist(figsize=(20,30))
```

```python
sns.boxplot(x="vs",y="am",data=car)
```

```python
sns.pairplot(car)
```

```python
car=car.drop(["hp","wt","vs"],axis=1)
car.head()
```

```python
car=car.rename(columns={"disp": "dp","am": "a"})#,"Unnamed: 0":"make"})
car.head(5)
```

```python
duplicate_row_car=car[car.duplicated()]
print("Number of duplicate row:",duplicate_row_car.shape)
```

```python
car.count()
```

```python
car=car.drop_duplicates()
car.head()
```

```python
print(car.isnull().sum())
```

```python
car=car.dropna()
car.count()
```

```python
car.cyl.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("number of car by mpg")
plt.ylabel("number of cars")
plt.xlabel("mpg")
```

```python
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
