```python
#calculating mean without using built_in functions
#Example no:1
from matplotlib import pyplot as plt
n_num = [1, 2, 3, 4, 5]
n = len(n_num)
get_sum = sum(n_num)
mean = get_sum / n
print("Mean / Average is: " +str(mean))
plt.hist(mean, bins=25, density=True, alpha=0.6, color='b')  
plt.show()
```

```python
#calculating median without using built_in functions
n_num = [4,8,6,9,5,5,6,7]
n = len(n_num)
n_num.sort()
if n % 2 == 0:
    median1 = n_num[n//2]
    median2 = n_num[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = n_num[n//2]
print("Median is: " + str(median))
plt.hist(median, bins=25, density=True, alpha=0.6, color='b')  
plt.show()
```

```python
#calculating mode without using built_in functions
from collections import Counter
n_num = [4,8,5,6,4,4,8,8]
n = len(n_num)
data = Counter(n_num)
get_mode = dict(data)
mode = [k for k, v in get_mode.items() if v == max(list(data.values()))]
if len(mode) == n:
    get_mode = "No mode found"
else:
    get_mode = "Mode is / are: " + ', '.join(map(str, mode))    
print(get_mode)
plt.hist(mode, bins=25, density=True, alpha=0.6, color='b')  
plt.show()
```

```python
#calculating variance and standard deviation without using library functions
import math
from matplotlib import pyplot as plt
xs = [0.5,0.7,0.3,0.2]     # values (must be floats!)
mean = sum(xs) / len(xs)   # mean
var  = sum(pow(x-mean,2) for x in xs) / len(xs)  # variance
std  = math.sqrt(var)  # standard deviation
print("Variance :",var)
print("Standard Deviation:",std)
plt.boxplot(xs)
plt.title("Boxplot")
plt.xlabel("data")
plt.ylabel("values")
plt.show()
```

```python
#calculating covariance without using library functions
import math
def mean(arr, n):
    sum = 0
    for i in range(0, n):
        sum = sum + arr[i]
    return sum / n
def covariance(arr1, arr2, n):
    sum = 0
    mean_arr1 = mean(arr1, n)
    mean_arr2 = mean(arr2, n)
    for i in range(0, n):
        sum = (sum + (arr1[i] - mean_arr1) * (arr2[i] - mean_arr2))
    return sum / (n - 1)
arr1 = [65.21, 64.75, 65.26, 65.76, 65.96]
n = len(arr1)
arr2 = [67.25, 66.39, 66.12, 65.70, 66.64]
m = len(arr2)
if (m == n):
    print("Co-variance ;",covariance(arr1, arr2, m))
plt.boxplot(arr1)
plt.title("Boxplot")
plt.xlabel("data")
plt.ylabel("values")
plt.show()
plt.boxplot(arr2)
plt.title("Boxplot")
plt.xlabel("data")
plt.ylabel("values")
plt.show()
```
